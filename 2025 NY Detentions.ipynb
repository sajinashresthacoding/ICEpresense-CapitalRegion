### In this notebook I will:
pull data from detentions the year 2025 for NY

```python
## import necessary libraries
import pandas as pd
```

```python
from datetime import datetime
```

```python
df1 = pd.read_excel('/Users/sajinashrestha/Documents/dataProjects/WAMC/ICEpresense-CapitalRegion/RawData/Detentions/2026-ICLI-00005_Detentions_FY25_20260311_Redacted.xlsx')
df1
```

```python
df1 = df1.iloc[5:]
df1
```

```python
df1.columns = df1.iloc[0]
df1_titled = df1[1:].reset_index(drop=True)
df1_titled
```

```python
df1_titled['Stay Book In Date Time'] = pd.to_datetime(df1_titled['Stay Book In Date Time'])
```

```python
dfDetentions1 = df1_titled.loc[
  (df1_titled["Stay Book In Date Time"] >= "2025-01-01") &
  (df1_titled["Stay Book In Date Time"] <= "2025-12-31")
]
```

```python
dfDetentions1
```

```python
dfDetentions1.to_csv('FilteredData/2025ICEDetentions1.csv', index=False)
```

```python
df2 = pd.read_excel('/Users/sajinashrestha/Documents/dataProjects/WAMC/ICEpresense-CapitalRegion/RawData/Detentions/2026-ICLI-00005_Detentions_FY26_20260311_Redacted.xlsx')
df2
```

```python
df2 = df2.iloc[5:]
df2
```

```python
df2.columns = df2.iloc[0]
df2_titled = df2[1:].reset_index(drop=True)
df2_titled
```

```python
df2_titled['Stay Book In Date Time'] = pd.to_datetime(df2_titled['Stay Book In Date Time'])
```

```python
dfDetentions2 = df2_titled.loc[
  (df2_titled["Stay Book In Date Time"] >= "2025-01-01") &
  (df2_titled["Stay Book In Date Time"] <= "2025-12-31")
]
```

```python
dfDetentions2
```

```python
dfDetentions2.to_csv('FilteredData/2025ICEDetentions2.csv', index=False)
```

```python
pdList = [dfDetentions1, dfDetentions2]
df_Detentions2025 = pd.concat(pdList).drop_duplicates()
df_Detentions2025
```
