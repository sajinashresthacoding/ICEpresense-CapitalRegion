```python
from datetime import datetime
```

```python
df["stay_book_in_date_time"] = pd.to_datetime(df["stay_book_in_date_time"])
```

```python
##siphoning just detentions made in the year of 2025

dfDetentions = df.loc[
  (df["stay_book_in_date_time"] >= "2025-01-01") &
  (df["stay_book_in_date_time"] <= "2025-12-31")] 
```

```python
dfDetentions
```

```python
dfArrestIDs = pd.read_csv('/Users/sajinashrestha/Documents/dataProjects/WAMC/ICEpresense-CapitalRegion/FilteredData/IDsNY2025Arrests_062326.csv')
dfArrestIDs
```

```python
#Now to sue the arrests identifier to only pull the detentions for arrests made in ny

filtered_detentions = dfDetentions[dfDetentions['unique_identifier'].isin(dfArrestIDs['unique_identifier'])]
filtered_detentions
```
