```python
#imports
import pandas as pd
import numpy as np
```

```python
arrests = pd.read_csv('/Users/sajinashrestha/Documents/dataProjects/WAMC/ICEpresense-CapitalRegion/FilteredData/All2025ICEArrests.csv')
arrests.info()
```

```python
detentions = pd.read_csv('/Users/sajinashrestha/Documents/dataProjects/WAMC/ICEpresense-CapitalRegion/FilteredData/All2025ICEDetentions.csv')
detentions.info()
```

```python
print(detentions['Detention Facility'].unique()) 
```

```python
#these are the nyfacilities thus far
## i checked these facilities by double checkign the name on the ICE website and seeing there were no duplicates
##for ex. multiple buffalo names but they all refer to the buffalo facility
##BUT: there are other clinton county jails so it was important to note WHICH clinton county i was talking about
my_facilities = [
    'BROOME COUNTY JAIL',
    'BATAVIA COMMAND CENTER HOLDROOM',
    'BUFFALO SPC',
    'BUFFALO USBP HOLD ROOM',
    'BUFFALO HOLD ROOM',
    'CLINTON COUNTY JAIL',
    'BROOKLYN MDC', 
    'ORANGE COUNTY JAIL',
    'ORANGE COUNT INTAKE RELEASE FAC.'    
]
```

```python
##now to extract only ny from the detention list
ny_detain = detentions[detentions['Detention Facility'].isin(my_facilities)]
ny_detain.info()
```

```python
ny_detain.to_csv('FilteredData/NYDetentions.csv')
```

```python
print(detentions['MSC Crime Class'].unique())
```

```python
##following sandras code, now to add charges to the arrests? 
```

```python
#get a table with the unique identifer and the charge
charge_lookup = detentions[['Anonymized Identifier', 'MSC Crime Class']].drop_duplicates(subset='Anonymized Identifier')
charge_lookup.info()
```

```python
#adds most serious conviction information, if present, to the arrests table
arrests_with_charge = arrests.merge(charge_lookup, on='Anonymized Identifier', how='left')
arrests_with_charge.info()
```

```python
arrests_with_charge.to_csv('Filtereddata/NYarrests_charges.csv')
```

```python
ny_removals = pd.read_csv('/Users/sajinashrestha/Documents/dataProjects/WAMC/ICEpresense-CapitalRegion/FilteredData/All2025ICERemovals.csv')
ny_removals.info()
```

```python
dfs = [arrests_with_charge, ny_detain, ny_removals]
```

```python
##let's try to group things?
import functools as ft
df_final = ft.reduce(lambda left, right: pd.merge(left, right, on='Anonymized Identifier'), dfs).drop_duplicates()
```

```python
df_final.to_csv("FilteredData/NYwholeJourney.csv")
```
